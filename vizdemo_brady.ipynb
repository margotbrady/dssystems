```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2024/2024-08-06/olympics.csv'
olympics = pd.read_csv(url)

# Display the first few rows
olympics.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.scatter(olympics.height, olympics.weight,s = 10)
plt.title('Olympian Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```

```python
# Import Seaborn
# Import Seaborn and Matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset
iris =  pd.read_csv('iris_dataset.csv')

# Compute the correlation matrix using only numeric columns
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_iris.corr()

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap of Numeric Features in Iris Dataset')
plt.show()
```
